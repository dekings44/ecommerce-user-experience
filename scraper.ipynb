```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the category page
url = 'https://www.konga.com/phones-tablets'

# Send an HTTP GET request
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

print(soup)
# Find all product items on the page
product_items = soup.find_all('div', class_='_4941f_1HCZm')

# Initialize lists to store scraped data
names = []
#descriptions = []
prices = []
#brands = []
sellers = []

# Iterate through product items and scrape data
for item in product_items:
    names.append(item.find('a', class_='af885_1iPzH').text.strip())
    #descriptions.append(item.find('div', class_='product-block-description').text.strip())
    prices.append(item.find('span', class_='d7c0f_sJAqi').text.strip())
    #brands.append(item.find('div', class_='_7cc7b_23GsY').text.strip())
    sellers.append(item.find('div', class_='_7cc7b_23GsY').text.strip())

# Create a DataFrame
data = {
    'Name': names,
    #'Description': descriptions,
    'Price': prices,
    #'Brand': brands,
    'Sold By': sellers
}
df = pd.DataFrame(data)

# Print or manipulate the DataFrame as needed
print(df)
```
